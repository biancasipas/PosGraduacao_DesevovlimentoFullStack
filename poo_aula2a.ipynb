### Exemplo de código 1: Definição de Classe e instanciação dos objetos

```python
# Definição da classe
class Aluno:

    def __init__(self, nome="Zé", idade=18): #Esse método construtor é chamado automaticamente quando um objeto da classe é criado
        self.nome = nome #Criando um atributo chamado nome.
        self.idade = idade

    def dobra_idade(self):
        return self.idade * 2


# Instanciação dos objetos

aluno_1 = Aluno()
aluno_1.nome = "Maria" # definindo que o nome do aluno_1 é "Maria"


aluno_2 = Aluno()
aluno_2.idade = 21 # definindo que a idade do aluno_2 é 21

print(aluno_1.nome) # imprime Maria (valor que definimos)
print(aluno_2.nome) # imprime Zé (valor default)

print(aluno_1.dobra_idade()) # dobra a idade 18 (valor default)
print(aluno_2.dobra_idade()) # dobra a idade 21 (valor que definimos)

# ao mudar o valor de nome em aluno_1, o aluno_2 continua inalterado
aluno_1.nome = "Ana"
print("Nome do aluno_1:", aluno_1.nome)
print("Nome do aluno_2:", aluno_2.nome)
```

# Oque é self ?
- `self` representa o próprio objeto da classe.
- Ele permite acessar os atributos e métodos daquele objeto.
- É parecido com o `this` do Java.
- Sem `self`, o Python não sabe de qual objeto você está falando.


```python
class Aluno:
    
    def __init__(self, nome="Sem nome", idade=0, nota=0.0, aprovado=False):
        self.nome = nome 
        self.idade = idade
        self.nota = nota
        self.aprovado = aprovado

    def fazer_aniversario(self):
        self.idade += 1
    
    def atualizar_nota(self,nova_nota):
        self.nota = nova_nota
    
    def verifica_aprovacao(self):
        if self.nota < 4:
            self.aprovado = False
            print("Reprovada")
        elif self.nota < 6:
            self.aprovado = False
            print("Media")
        else:
            self.aprovado = True
            print("Aprovada")
    
    def mostra_dados(self):
        print("Nome:", self.nome)
        print("Idade:", self.idade)
        print("Nota:", self.nota)
        print("Aprovado:", self.aprovado)

    
    aluno1 = Aluno("Maria", 20)
    aluno1.atualizar_nota(7.5)
    aluno1.verifica_aprovacao()
    aluno1.fazer_aniversario()
    aluno1.mostra_dados()

    print("----------")

    aluno2 = Aluno("João", 18)
    aluno2.atualizar_nota(3.0)
    aluno2.verifica_aprovacao()
    aluno2.fazer_aniversario()
    aluno2.mostra_dados()

    #INPUT
    #nome = input("Digite o nome do aluno: ")
    #idade = int(input("Digite a idade: "))
    #nota = float(input("Digite a nota: "))

    #aluno = Aluno(nome, idade, nota)
    #aluno.verifica_aprovacao()
    #aluno.fazer_aniversario()
    #aluno.mostra_dados()



    
    
```

### Exemplo de código 2: Parâmetro self

```python
class Calculadora:

    def __init__(self):
        self.valor_interno = 5

    def adiciona(self, numero):
        # não podemos chamar a variável "valor_interno" sem "self" porque "valor" é um atributo de Calculadora
        return self.valor_interno + numero

total = Calculadora()
print("resultado = %d" % total.adiciona(7)) # self pegará o valor 5 de valor_interno e o valor 3 irá para numero
```

### Exemplo de código 3: Método \_\_init__

```python
class Carro:

    def __init__(self):
        self.cor = "preto"

# um novo carro é criado com cor = "preto"
carro_1 = Carro()
print(carro_1.cor)

# alterando o valor da color de carro_1 para "branco"
carro_1.cor = "branco"
print(carro_1.cor)
```

```python
class Carro2:

    def __init__(self, cor):
        self.cor = cor

# um novo carro é criado com cor = "azul"
carro_2 = Carro2("azul")
print(carro_2.cor)

# alterando o valor da color de carro_2 para "branco"
carro_2.cor = "branco"
print(carro_2.cor)
```
